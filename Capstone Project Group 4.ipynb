```python



import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
from scipy import stats
import warnings

warnings.filterwarnings("ignore")

# ── 0. Style ─────────────────────────────────────────────────────────────────
sns.set_theme(style="whitegrid", palette="muted", font_scale=1.05)
plt.rcParams.update({"figure.dpi": 130, "axes.spines.top": False,
                     "axes.spines.right": False})

# DATA_PATH = "Wholesale_customers_data_1_.csv"   # adjust if needed
FEATURES  = ["fresh", "milk", "grocery", "frozen", "detergents_paper", "delicassen"]

CHANNEL_MAP = {1: "HoReCa", 2: "Retail"}
REGION_MAP  = {1: "Lisbon", 2: "Oporto", 3: "Other"}
PALETTE_CH  = {"HoReCa": "#378ADD", "Retail": "#1D9E75"}
PALETTE_REG = {"Lisbon": "#D85A30", "Oporto": "#D4537E", "Other": "#534AB7"}


# ==============================================================================
# 1. LOAD & BASIC INSPECTION
# ==============================================================================

df = data.copy() # Use the already loaded 'data' DataFrame
df["channel_label"] = df["channel"].map(CHANNEL_MAP)
df["region_label"]  = df["region"].map(REGION_MAP)

print("=" * 60)
print("1. BASIC INFORMATION")
print("=" * 60)
print(f"\nShape : {df.shape}  ({df.shape[0]} rows, {df.shape[1]} columns)")
print(f"\nData types:\n{df.dtypes}")
print(f"\nMissing values:\n{df.isnull().sum()}")
print(f"\nDuplicates : {df.duplicated().sum()}")


# ==============================================================================
# 2. DESCRIPTIVE STATISTICS
# ==============================================================================

print("\n" + "=" * 60)
print("2. DESCRIPTIVE STATISTICS")
print("=" * 60)

desc = df[FEATURES].describe().T
desc["skewness"] = df[FEATURES].skew()
desc["kurtosis"] = df[FEATURES].kurt()
desc["cv"]       = desc["std"] / desc["mean"]        # coefficient of variation

print(f"\n{desc.round(2).to_string()}")


# ==============================================================================
# 3. CATEGORICAL DISTRIBUTIONS
# ==============================================================================

print("\n" + "=" * 60)
print("3. CATEGORICAL DISTRIBUTIONS")
print("=" * 60)

ch_counts  = df["channel_label"].value_counts()
reg_counts = df["region_label"].value_counts()

print(f"\nChannel:\n{ch_counts.to_string()}")
print(f"\nRegion:\n{reg_counts.to_string()}")

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
fig.suptitle("Categorical Feature Distributions", fontweight="bold")

axes[0].pie(
    ch_counts.values,
    labels=[f"{l}\n({v})" for l, v in zip(ch_counts.index, ch_counts.values)],
    colors=[PALETTE_CH[k] for k in ch_counts.index],
    autopct="%1.1f%%", startangle=90,
    wedgeprops={"linewidth": 1, "edgecolor": "white"}
)
axes[0].set_title("Channel")

axes[1].pie(
    reg_counts.values,
    labels=[f"{l}\n({v})" for l, v in zip(reg_counts.index, reg_counts.values)],
    colors=[PALETTE_REG[k] for k in reg_counts.index],
    autopct="%1.1f%%", startangle=90,
    wedgeprops={"linewidth": 1, "edgecolor": "white"}
)
axes[1].set_title("Region")

plt.tight_layout()
plt.savefig("01_categorical_distributions.png", bbox_inches="tight")
plt.show()
print("  → saved 01_categorical_distributions.png")


# ==============================================================================
# 4. UNIVARIATE DISTRIBUTIONS (histograms + KDE + box)
# ==============================================================================

print("\n" + "=" * 60)
print("4. UNIVARIATE DISTRIBUTIONS")
print("=" * 60)

fig, axes = plt.subplots(3, 6, figsize=(18, 10))
fig.suptitle("Univariate Distributions — Raw & Log-transformed", fontweight="bold")

for i, col in enumerate(FEATURES):
    ax_raw = axes[0, i]
    ax_log = axes[1, i]
    ax_box = axes[2, i]

    # Raw histogram + KDE
    sns.histplot(df[col], kde=True, ax=ax_raw, color="#378ADD", linewidth=0)
    ax_raw.set_title(col, fontsize=10)
    ax_raw.set_xlabel("")
    ax_raw.tick_params(labelsize=8)

    # Log-transformed
    log_data = np.log1p(df[col])
    sns.histplot(log_data, kde=True, ax=ax_log, color="#1D9E75", linewidth=0)
    ax_log.set_title(f"log({col})", fontsize=10)
    ax_log.set_xlabel("")
    ax_log.tick_params(labelsize=8)

    # Box plot
    ax_box.boxplot(df[col], vert=True, patch_artist=True,
                   boxprops=dict(facecolor="#B5D4F4", color="#185FA5"),
                   medianprops=dict(color="#185FA5", linewidth=2),
                   whiskerprops=dict(color="#888"),
                   capprops=dict(color="#888"),
                   flierprops=dict(marker="o", color="#D85A30", alpha=0.5, markersize=4))
    ax_box.set_xlabel("")
    ax_box.tick_params(labelsize=8)

for row, label in enumerate(["Raw", "Log1p", "Boxplot"]):
    axes[row, 0].set_ylabel(label, fontsize=9)

plt.tight_layout()
plt.savefig("02_univariate_distributions.png", bbox_inches="tight")
plt.show()
print("  → saved 02_univariate_distributions.png")


# ==============================================================================
# 5. OUTLIER ANALYSIS
# ==============================================================================

print("\n" + "=" * 60)
print("5. OUTLIER ANALYSIS (IQR method)")
print("=" * 60)

outlier_summary = []
for col in FEATURES:
    Q1, Q3 = df[col].quantile([0.25, 0.75])
    IQR    = Q3 - Q1
    lo, hi = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    mask   = (df[col] < lo) | (df[col] > hi)
    outlier_summary.append({
        "Feature": col, "Q1": Q1, "Q3": Q3, "IQR": IQR,
        "Lower fence": lo, "Upper fence": hi, "Outlier count": mask.sum(),
        "Outlier %": round(mask.mean() * 100, 2)
    })

out_df = pd.DataFrame(outlier_summary).set_index("Feature")
print(f"\n{out_df.round(2).to_string()}")

fig, ax = plt.subplots(figsize=(9, 4))
colors = ["#D85A30" if v > 35 else "#378ADD" for v in out_df["Outlier count"]]
bars = ax.bar(out_df.index, out_df["Outlier count"], color=colors, width=0.5)
ax.bar_label(bars, padding=3, fontsize=10)
ax.set_title("Outlier Count per Feature (IQR method)", fontweight="bold")
ax.set_ylabel("Outlier count")
ax.set_xlabel("")
plt.tight_layout()
plt.savefig("03_outlier_counts.png", bbox_inches="tight")
plt.show()
print("  → saved 03_outlier_counts.png")


# ==============================================================================
# 6. CORRELATION ANALYSIS
# ==============================================================================

print("\n" + "=" * 60)
print("6. CORRELATION ANALYSIS")
print("=" * 60)

corr_matrix = df[FEATURES].corr()
print(f"\nPearson correlation matrix:\n{corr_matrix.round(3).to_string()}")

# Heatmap
fig, ax = plt.subplots(figsize=(8, 6))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool), k=1)
sns.heatmap(
    corr_matrix, annot=True, fmt=".2f", cmap="coolwarm",
    vmin=-1, vmax=1, linewidths=0.5, ax=ax, square=True,
    cbar_kws={"shrink": 0.8}
)
ax.set_title("Correlation Heatmap", fontweight="bold")
plt.tight_layout()
plt.savefig("04_correlation_heatmap.png", bbox_inches="tight")
plt.show()
print("  → saved 04_correlation_heatmap.png")

# Highlight strong correlations
print("\nStrong correlations (|r| > 0.5):")
for i, f1 in enumerate(FEATURES):
    for j, f2 in enumerate(FEATURES):
        if j > i and abs(corr_matrix.loc[f1, f2]) > 0.5:
            print(f"  {f1} ↔ {f2} : r = {corr_matrix.loc[f1, f2]:.3f}")


# ==============================================================================
# 7. BIVARIATE — PAIRPLOT
# ==============================================================================

print("\n" + "=" * 60)
print("7. PAIRPLOT (log scale, coloured by Channel)")
print("=" * 60)

df_log = df[FEATURES + ["channel_label"]].copy()
for col in FEATURES:
    df_log[col] = np.log1p(df[col])

g = sns.pairplot(
    df_log, hue="channel_label", palette=PALETTE_CH,
    plot_kws={"alpha": 0.4, "s": 20},
    diag_kind="kde", corner=True
)
g.figure.suptitle("Pairplot (log-transformed, coloured by Channel)", y=1.01, fontweight="bold")
g.figure.savefig("05_pairplot.png", bbox_inches="tight")
plt.show()
print("  → saved 05_pairplot.png")


# ==============================================================================
# 8. CHANNEL-WISE ANALYSIS
# ==============================================================================

print("\n" + "=" * 60)
print("8. CHANNEL-WISE ANALYSIS")
print("=" * 60)

ch_means = df.groupby("channel_label")[FEATURES].mean().round(2)
print(f"\nMean spend by channel:\n{ch_means.to_string()}")

# Mann-Whitney U test (non-parametric, given skewness)
print("\nMann-Whitney U test (HoReCa vs Retail) per feature:")
hoReCa = df[df["channel"] == 1]
retail = df[df["channel"] == 2]
for col in FEATURES:
    stat, p = stats.mannwhitneyu(hoReCa[col], retail[col], alternative="two-sided")
    sig = "***" if p < 0.001 else ("**" if p < 0.01 else ("*" if p < 0.05 else "ns"))
    print(f"  {col:<20} U={stat:.0f}  p={p:.4f}  {sig}")

# Grouped bar chart
ch_means_t = ch_means.T
fig, ax = plt.subplots(figsize=(10, 5))
x  = np.arange(len(FEATURES))
w  = 0.35
b1 = ax.bar(x - w/2, ch_means_t["HoReCa"], w, label="HoReCa", color="#378ADD")
b2 = ax.bar(x + w/2, ch_means_t["Retail"],  w, label="Retail",  color="#1D9E75")
ax.set_xticks(x)
ax.set_xticklabels(FEATURES, rotation=20, ha="right")
ax.set_title("Mean Spend by Channel", fontweight="bold")
ax.set_ylabel("Mean annual spend")
ax.legend()
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda v, _: f"{v/1000:.0f}k"))
plt.tight_layout()
plt.savefig("06_channel_comparison.png", bbox_inches="tight")
plt.show()
print("  → saved 06_channel_comparison.png")


# ==============================================================================
# 9. REGION-WISE ANALYSIS
# ==============================================================================

print("\n" + "=" * 60)
print("9. REGION-WISE ANALYSIS")
print("=" * 60)

reg_means = df.groupby("region_label")[FEATURES].mean().round(2)
print(f"\nMean spend by region:\n{reg_means.to_string()}")

fig, axes = plt.subplots(2, 3, figsize=(14, 8), sharey=False)
fig.suptitle("Spend Distribution by Region", fontweight="bold")

for i, col in enumerate(FEATURES):
    ax = axes[i // 3, i % 3]
    sns.boxplot(data=df, x="region_label", y=col, ax=ax,
                palette=PALETTE_REG, order=["Lisbon", "Oporto", "Other"],
                linewidth=0.8, fliersize=3)
    ax.set_title(col, fontsize=11)
    ax.set_xlabel("")
    ax.set_ylabel("")
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda v, _: f"{int(v/1000)}k" if v >= 1000 else str(int(v))))

plt.tight_layout()
plt.savefig("07_region_boxplots.png", bbox_inches="tight")
plt.show()
print("  → saved 07_region_boxplots.png")


# ==============================================================================
# 10. CROSS-TAB — CHANNEL × REGION
# ==============================================================================

print("\n" + "=" * 60)
print("10. CHANNEL × REGION CROSS-TAB")
print("=" * 60)

xt = pd.crosstab(df["channel_label"], df["region_label"],
                 margins=True, margins_name="Total")
print(f"\n{xt.to_string()}")

chi2, p, dof, expected = stats.chi2_contingency(
    pd.crosstab(df["channel_label"], df["region_label"]))
print(f"\nChi-square test: χ²={chi2:.3f}, df={dof}, p={p:.4f}")
if p < 0.05:
    print("  → Channel and Region are NOT independent (p < 0.05)")
else:
    print("  → No significant association between Channel and Region")


# ==============================================================================
# 11. SKEWNESS & NORMALITY TEST
# ==============================================================================

print("\n" + "=" * 60)
print("11. SKEWNESS & NORMALITY (Shapiro-Wilk, n=440)")
print("=" * 60)

for col in FEATURES:
    sk = df[col].skew()
    # Shapiro-Wilk limited to 5000 samples
    stat, p = stats.shapiro(df[col].sample(min(len(df), 200), random_state=42))
    print(f"  {col:<20} skew={sk:.3f}  W={stat:.4f}  p={p:.4e}  {'Non-normal' if p < 0.05 else 'Normal'}")


# ==============================================================================
# 12. SUMMARY DASHBOARD
# ==============================================================================

fig = plt.figure(figsize=(16, 10))
gs  = gridspec.GridSpec(2, 3, figure=fig, hspace=0.4, wspace=0.4)
fig.suptitle("Wholesale Customers — EDA Summary Dashboard", fontweight="bold", fontsize=14)

# (a) Pie — Channel
ax0 = fig.add_subplot(gs[0, 0])
ax0.pie([298, 142], labels=["HoReCa\n298", "Retail\n142"],
        colors=["#378ADD", "#1D9E75"], autopct="%1.1f%%",
        wedgeprops={"edgecolor": "white"})
ax0.set_title("Channel split")

# (b) Pie — Region
ax1 = fig.add_subplot(gs[0, 1])
ax1.pie([316, 77, 47], labels=["Other\n316", "Lisbon\n77", "Oporto\n47"],
        colors=["#534AB7", "#D85A30", "#D4537E"], autopct="%1.1f%%",
        wedgeprops={"edgecolor": "white"})
ax1.set_title("Region split")

# (c) Outlier bar
ax2 = fig.add_subplot(gs[0, 2])
outlier_counts = [20, 28, 24, 43, 30, 27]
colors2 = ["#D85A30" if v > 35 else "#378ADD" for v in outlier_counts]
bars2 = ax2.bar(FEATURES, outlier_counts, color=colors2)
ax2.bar_label(bars2, padding=2, fontsize=9)
ax2.set_title("Outliers per feature (IQR)")
ax2.set_xticklabels(FEATURES, rotation=25, ha="right", fontsize=8)

# (d) Correlation heatmap
ax3 = fig.add_subplot(gs[1, :2])
sns.heatmap(df[FEATURES].corr(), annot=True, fmt=".2f", cmap="coolwarm",
            vmin=-1, vmax=1, linewidths=0.5,
            cbar_kws={"shrink": 0.8}, annot_kws={"size": 9})
ax3.set_title("Correlation matrix")
ax3.tick_params(labelsize=8)

# (e) Skewness bar
ax4 = fig.add_subplot(gs[1, 2])
skews = df[FEATURES].skew().values
bar_colors = ["#D85A30" if s > 5 else "#378ADD" for s in skews]
ax4.barh(FEATURES, skews, color=bar_colors)
ax4.axvline(0, color="#888", linewidth=0.8)
ax4.set_title("Skewness")
ax4.tick_params(labelsize=9)

plt.savefig("08_summary_dashboard.png", bbox_inches="tight")
plt.show()



```

```python
import numpy as np
import pandas as pd

from sklearn.cluster import (
    KMeans,
    AgglomerativeClustering,
    DBSCAN
)

from sklearn.neighbors import NearestNeighbors

from sklearn.metrics import (
    silhouette_score,
    davies_bouldin_score,
    calinski_harabasz_score,
)

from sklearn.preprocessing import StandardScaler


# ─────────────────────────────────────────────
# K-MEANS TUNING
# ─────────────────────────────────────────────

def tune_kmeans(X_scaled, k_range, random_state=42):

    records = []

    for k in k_range:

        model = KMeans(
            n_clusters=k,
            n_init=20,
            random_state=random_state
        )

        labels = model.fit_predict(X_scaled)

        records.append({
            "k": k,
            "inertia": model.inertia_,
            "silhouette": silhouette_score(X_scaled, labels),
            "davies_bouldin": davies_bouldin_score(X_scaled, labels),
            "calinski_harabasz": calinski_harabasz_score(X_scaled, labels),
        })

    return pd.DataFrame(records)


# ─────────────────────────────────────────────
# AGGLOMERATIVE TUNING
# ─────────────────────────────────────────────

def tune_agglomerative(X_scaled, cluster_range, linkage="ward"):

    records = []

    for n in cluster_range:

        model = AgglomerativeClustering(
            n_clusters=n,
            linkage=linkage
        )

        labels = model.fit_predict(X_scaled)

        records.append({
            "n_clusters": n,
            "linkage": linkage,
            "silhouette": silhouette_score(X_scaled, labels),
            "davies_bouldin": davies_bouldin_score(X_scaled, labels),
            "calinski_harabasz": calinski_harabasz_score(X_scaled, labels),
        })

    return pd.DataFrame(records)


# ─────────────────────────────────────────────
# K-DISTANCE FOR DBSCAN
# ─────────────────────────────────────────────

def compute_k_distance(X_scaled, min_samples=5):

    nn = NearestNeighbors(n_neighbors=min_samples)

    nn.fit(X_scaled)

    distances, _ = nn.kneighbors(X_scaled)

    k_distances = distances[:, min_samples - 1]

    return np.sort(k_distances)


# ─────────────────────────────────────────────
# DBSCAN TUNING
# ─────────────────────────────────────────────

def tune_dbscan(X_scaled, param_grid):

    records = []

    for eps, min_samples in param_grid:

        model = DBSCAN(
            eps=eps,
            min_samples=min_samples
        )

        labels = model.fit_predict(X_scaled)

        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)

        noise_ratio = (labels == -1).mean()

        # Remove noise points for evaluation
        mask = labels != -1

        filtered_labels = labels[mask]
        filtered_X = X_scaled[mask]

        unique_clusters = len(set(filtered_labels))

        if unique_clusters >= 2:

            sil = silhouette_score(filtered_X, filtered_labels)

            dbi = davies_bouldin_score(
                filtered_X,
                filtered_labels
            )

            chi = calinski_harabasz_score(
                filtered_X,
                filtered_labels
            )

        else:
            sil = np.nan
            dbi = np.nan
            chi = np.nan

        records.append({
            "eps": eps,
            "min_samples": min_samples,
            "clusters": n_clusters,
            "noise_ratio": noise_ratio,
            "silhouette": sil,
            "davies_bouldin": dbi,
            "calinski_harabasz": chi,
        })

    return pd.DataFrame(records)


# ─────────────────────────────────────────────
# MAIN PROGRAM
# ─────────────────────────────────────────────

if __name__ == "__main__":

    # Load dataset from Google Drive

    url = "https://drive.google.com/file/d/1yDhLUJoB-Z69OL2Ad7tvm_0Gt6Dh2a5o/view?usp=drive_link"

    url = "https://drive.google.com/uc?id=" + url.split("/")[-2]

    df = pd.read_csv(url)

    # Feature selection

    FEATURES = [
        "Fresh",
        "Milk",
        "Grocery",
        "Frozen",
        "Detergents_Paper",
        "Delicassen"
    ]

    X_raw = df[FEATURES].copy()

    # Log transform

    X_log = np.log1p(X_raw)

    # Standardization

    scaler = StandardScaler()

    X_scaled = scaler.fit_transform(X_log)

    # ─────────────────────────────────────────
    # K-MEANS
    # ─────────────────────────────────────────

    print("\n── K-Means Tuning ──")

    km_results = tune_kmeans(
        X_scaled,
        k_range=range(2, 11)
    )

    print(km_results.to_string(index=False))

    best_k = int(
        km_results.loc[
            km_results["silhouette"].idxmax(),
            "k"
        ]
    )

    print(f"\nBest k: {best_k}")

    # ─────────────────────────────────────────
    # AGGLOMERATIVE
    # ─────────────────────────────────────────

    print("\n── Agglomerative Tuning ──")

    agg_results = tune_agglomerative(
        X_scaled,
        cluster_range=range(2, 8)
    )

    print(agg_results.to_string(index=False))

    # ─────────────────────────────────────────
    # DBSCAN
    # ─────────────────────────────────────────

    min_samples = len(FEATURES) + 1

    k_dist = compute_k_distance(
        X_scaled,
        min_samples=min_samples
    )

    eps_candidates = np.percentile(
        k_dist,
        [50, 60, 70, 75, 80]
    )

    print("\nSuggested eps values:")

    print(np.round(eps_candidates, 3))

    param_grid = [
        (round(eps, 3), min_samples)
        for eps in eps_candidates
    ]

    db_results = tune_dbscan(
        X_scaled,
        param_grid
    )

    print("\n── DBSCAN Tuning ──")

    print(db_results.to_string(index=False))

    # ─────────────────────────────────────────
    # FINAL COMPARISON
    # ─────────────────────────────────────────

    print("\n── Comparative Summary ──")

    best_km = km_results.loc[
        km_results["silhouette"].idxmax()
    ]

    best_agg = agg_results.loc[
        agg_results["silhouette"].idxmax()
    ]

    best_db = db_results.dropna(
        subset=["silhouette"]
    )

    if not best_db.empty:
        best_db = best_db.loc[
            best_db["silhouette"].idxmax()
        ]

    summary_rows = [

        {
            "Algorithm": f"K-Means (k={int(best_km['k'])})",
            "Silhouette": round(best_km["silhouette"], 4),
            "Davies-Bouldin": round(best_km["davies_bouldin"], 4),
            "Calinski-Harabasz": round(best_km["calinski_harabasz"], 2),
        },

        {
            "Algorithm": f"Agglomerative (k={int(best_agg['n_clusters'])})",
            "Silhouette": round(best_agg["silhouette"], 4),
            "Davies-Bouldin": round(best_agg["davies_bouldin"], 4),
            "Calinski-Harabasz": round(best_agg["calinski_harabasz"], 2),
        }
    ]

    if not best_db.empty:

        summary_rows.append({

            "Algorithm":
            f"DBSCAN (eps={best_db['eps']}, "
            f"min_samples={int(best_db['min_samples'])})",

            "Silhouette": round(best_db["silhouette"], 4),

            "Davies-Bouldin":
            round(best_db["davies_bouldin"], 4),

            "Calinski-Harabasz":
            round(best_db["calinski_harabasz"], 2),
        })

    summary = pd.DataFrame(summary_rows)

    print(summary.to_string(index=False))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import davies_bouldin_score, silhouette_score
from sklearn.preprocessing import StandardScaler


# ─────────────────────────────
# LOAD DATA
# ─────────────────────────────

url = "https://drive.google.com/file/d/1yDhLUJoB-Z69OL2Ad7tvm_0Gt6Dh2a5o/view?usp=drive_link"
url = "https://drive.google.com/uc?id=" + url.split("/")[-2]

df = pd.read_csv(url)

FEATURES = [
    "Fresh", "Milk", "Grocery",
    "Frozen", "Detergents_Paper", "Delicassen"
]

X = np.log1p(df[FEATURES])
X_scaled = StandardScaler().fit_transform(X)


# ─────────────────────────────
# K-MEANS
# ─────────────────────────────

k_vals = range(2, 11)
inertias = []
dbis = []

for k in k_vals:

    model = KMeans(
        n_clusters=k,
        n_init=5,
        random_state=42
    )

    labels = model.fit_predict(X_scaled)

    inertias.append(model.inertia_)
    dbis.append(davies_bouldin_score(X_scaled, labels))


# ─────────────────────────────
# PLOT: ELBOW + DBI
# ─────────────────────────────

plt.figure()
plt.plot(k_vals, inertias, marker='o')
plt.title("Elbow Method (K-Means Inertia)")
plt.xlabel("k")
plt.ylabel("Inertia")
plt.grid()
plt.show()

plt.figure()
plt.plot(k_vals, dbis, marker='o')
plt.title("Davies-Bouldin Index vs k")
plt.xlabel("k")
plt.ylabel("DBI (lower is better)")
plt.grid()
plt.show()


# ─────────────────────────────
# BEST K-MEANS + SILHOUETTE
# ─────────────────────────────

best_k = k_vals[np.argmin(dbis)]

best_model = KMeans(
    n_clusters=best_k,
    n_init=5,
    random_state=42
)

labels = best_model.fit_predict(X_scaled)

sil = silhouette_score(
    X_scaled,
    labels,
    sample_size=500
)

print("\nBest k:", best_k)
print("Silhouette:", sil)


# ─────────────────────────────
# QUICK CLUSTER VISUAL (PCA REDUCTION)
# ─────────────────────────────

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_2d = pca.fit_transform(X_scaled)

plt.figure()
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=labels, cmap="viridis", s=20)
plt.title("K-Means Clusters (PCA Projection)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
